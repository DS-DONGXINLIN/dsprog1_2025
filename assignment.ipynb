```python
import requests
from bs4 import BeautifulSoup
import time
import urllib.parse

BASE_URL = "https://www.musashino-u.ac.jp/"

pages = {}

def is_same_domain(url):
    netloc = urllib.parse.urlparse(url).netloc
    return ("musashino-u.ac.jp" in netloc) or (netloc == "")

def abs_url(base, link):
    return urllib.parse.urljoin(base, link)
```

```python

def is_same_domain(url):
    domain = urllib.parse.urlparse(url).netloc
    return ("musashino-u.ac.jp" in domain) or (domain == "")

def get_absolute_url(base, link):
    return urllib.parse.urljoin(base, link)

# time.sleep(1)
res = requests.get(BASE_URL)
soup = BeautifulSoup(res.text, "html.parser")

if soup.title:
    pages[BASE_URL] = soup.title.get_text(strip=True)
else:
    pages[BASE_URL] = ""

links = soup.find_all("a", href=True)

for tag in links:
    href = tag["href"].strip()

    if href.startswith("#") or "#" in href:
        continue

    if href.lower().startswith("mailto:"):
        continue

    full = abs_url(BASE_URL, href)

    if not is_same_domain(full):
        continue

    # time.sleep(1)

    try:
        resp = requests.get(full)
        sp = BeautifulSoup(resp.text, "html.parser")
        t = sp.title.get_text(strip=True) if sp.title else ""
        pages[full] = t
    except:
        pages[full] = ""
```

```python
print("=== 抽出的URL与标题 ===")
for u, t in pages.items():
    print(u, " -> ", t)
```
